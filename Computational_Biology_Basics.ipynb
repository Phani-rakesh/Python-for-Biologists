```python
def count_low_bases(q_string, threshold=20):
    count = 0
    for char in q_string:
        q = ord(char) - 33
        if q < threshold:
            count += 1
    return count


cleaned_reads = []

for i in range(0, len(lines), 4):
    header = lines[i]
    sequence = lines[i+1]
    plus = lines[i+2]
    quality = lines[i+3]

    low_bases = count_low_bases(quality.strip())
    percent_low = (low_bases / len(quality.strip())) * 100

    if percent_low <= 20:
        cleaned_reads.extend([header, sequence, plus, quality])

with open("cleaned_patient.fastq", "w") as f:
    f.writelines(cleaned_reads)

print("Cleaned FASTQ created.")
```

```python
reference = "ATGACTGACCTGAGGAGAAGTCTGCCGTTACTG"
```

```python
def find_mutations(seq, ref):
    mutations = []
    for i in range(len(seq)):
        if seq[i] != ref[i]:
            mutations.append((i, ref[i], seq[i]))
    return mutations


report_data = []

for i in range(0, len(cleaned_reads), 4):
    header = cleaned_reads[i].strip()
    sequence = cleaned_reads[i+1].strip()
    quality = cleaned_reads[i+3].strip()

    mutations = find_mutations(sequence, reference)

    report_data.append({
        "header": header,
        "mutations": mutations
    })
```

```python
report_lines = []

report_lines.append("=== Cancer NGS Variant Report ===\n")
report_lines.append("Patient ID: CH-2026-017\n\n")

for entry in report_data:
    report_lines.append(f"{entry['header']}\n")

    if entry["mutations"]:
        for m in entry["mutations"]:
            report_lines.append(
                f"Mutation at position {m[0]}: {m[1]} -> {m[2]}\n"
            )
    else:
        report_lines.append("No mutation detected.\n")

    report_lines.append("\n")

with open("final_report.txt", "w") as f:
    f.writelines(report_lines)

print("Final report generated.")
```
